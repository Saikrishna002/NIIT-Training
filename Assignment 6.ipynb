```python
#1 Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2 and 3: Reading the file and naming it
wine = pd.read_csv('Wine.data', sep = ',', header = None)
wine
```

```python
# 4. Deleting the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns
wine.drop(columns = [0,3,6,8,10,12,13])
```

```python
#5 Assigning the columns as below:
wine.rename(columns = {0: 'Type', 1: 'Alcohol', 2: 'Malic Acid', 3: 'Ash', 4: 'Alcalinity of Ash', 5: 'Magnesium', 6: 'Total Phenols', 7:'Flavanoids', 8: 'Nonflavanoid phenols', 9:'Proanthocyanins', 10: 'Color intensity', 11:'Hue', 12:'OD280/OD315', 13: 'Proline'}, inplace = True)
wine
```

```python
# 6. Setting the values of the first 3 rows from alcohol as NaN
wine.iloc[0:3,1] = np.nan
wine
```

```python
#7 Now setting the value of the rows 3 and 4 of magnesium as NaN
wine.iloc[2:4,5] = np.nan
wine 
```

```python
#8 Filling the value of NaN with the number 10 in alcohol and 100 in magnesium
wine['Alcohol'].fillna(10, inplace = True )
wine['Magnesium'].fillna(100, inplace = True)
wine
```

```python
#Step 9. Counting the number of missing values
wine.isnull().sum()
```

```python
#Step 10. Creating an array of 10 random numbers up until 10
randnums= np.random.randint(0,10,10)
randnums
```

```python
#Step 11. Using random numbers you generated as an index and assign NaN value to each of cell.
wine.loc[randnums] = np.nan
wine.head(10)
```

```python
#Step 12. How many missing values do we have?
wine.isnull().sum()
```

```python
#Step 13. Deleting the rows that contain missing values
wine.dropna(inplace = True)
wine
```

```python
#Step 14. Printing only the non-null values in alcohol
wine.loc[wine['Alcohol'].notnull()]
```

```python
#Step 15. Reseting the index, so it starts with 0 again
wine.reset_index(inplace = True, drop = True)
wine
```
