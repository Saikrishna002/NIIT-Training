```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
#Importing the dataset from this address.Assign it to a variable called iris
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
```

```python
iris = df
iris.head()
```

```python
# Step 4. Creating columns for the dataset
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class

iris.columns =['sepal_length (in cm)', 'sepal_width (in cm)', 'petal_length (in cm)', 'petal_width (in cm)','class']
iris
```

```python
#Creating columns for the dataset
iris.rename(columns={0:'sepal_length (in cm)',1:'sepal_width (in cm)',2:'petal_length (in cm)',3:'petal_width (in cm)',4:'class'},inplace=True)
```

```python
#Is there any missing value in the dataframe?
iris.isnull().sum()
```

```python
#Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN
iris.loc[10:29, ['petal_length (in cm)']] = [np.nan]
iris.head(32)
```

```python
#Substituting the NaN values to 1.0
iris['petal_length (in cm)'].iloc[9:29].replace(np.nan,1.0,inplace=True)
iris['petal_length (in cm)'].iloc[9:29]
```

```python
#Deleting the index class
iris.drop('index',axis=1,inplace=True)
iris
```

```python
#Seting the first 3 rows as NaN
iris.iloc[0:3]=np.nan
iris
```

```python
#Deleting the rows that have NaN
iris.dropna(inplace=True)
```

```python
#Reseting the index so it begins with 0 again
iris.reset_index(inplace=True)
```
